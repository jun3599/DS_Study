### 머신러닝의 분류 

#### 지도학습
* 지도학습은 주어진 입력으로부터 출력을 예측하고자 할때 사용 
* input과 함께 정답 데이터를 제공하여 학습 
* 새로운 입력 데이터에 대해 정확한 출력을 예측하도록 하는것이 목표

##### 분류와 회귀
* 지도학습은 크게 **분류(classification)** 과 **회귀(Regression)** 으로 구분 
* 분류는 입력 데이터를 사전에 정의된 여러개의 클레스중 하나로 예측하는것이며, 클레스의 갯수에 따라 이진분류, 다중분류 등으로 나누어 불리기도 한다. 
* 회귀는 연속적인 숫자를 예측하는것


##### 지도학습 알고리즘 
* 선형회귀 (회귀)
* 로지스틱 회귀 (분류)
* 서포트벡터 머신 
* k-최근접 이웃
* 결정트리 
* 앙상블 
* 신경망

#### 비지도학습
* 비지도 학습은 원하는 출력 없이 입력 데이터를 사용 
* 입력 데이터의 구조나 패턴을 찾는것이 목표 
* 미리 정해진 결과가 없고, 방대한 양의 데이터에서 유용한 통찰력을 얻을 수 있음 

##### 클러스터링, 차원축소, 연관규칙 
* 비지도 학습은 크게 클러스터링(Clustering), 차원축소(Dimensionality Reduction), 연관규칙(Association Rules)으로 구분 
* 클러스터링은 공간상에 서로 가깝고 유사한 데이터를 클러스터로 그룹화
* 차원축소는 고차원의 데이터에 대해서 너무 많은 정보를 잃지 않으면서 데이터를 축소시키는 방법 
* 연관규칙은 데이터에서 특성간의 연관성이 있는 흥미로운 규칙을 찾는 방법

##### 비지도학습  알고리즘 
* 클러스터링
    * K-means  
    * DBSCAN: 밀도기반 클러스터링(Denscity-based spatial clustering of applications with noise)  
    https://bcho.tistory.com/1205  
    * 계층 군집 분석: 계층을 갖는 군집을 구성합니다.    
    * 이상치 탐지, 특이값 탐지  
* 차원 축소 
    * 주성분 분석(PCA)    
    * 커널 PCA   
    * t-SNE(t-Distributed Stochastic Neighbor Embedding) 
* 연관규칙 
    * Apriori 
    * Eclat

#### 준지도 학습
* 레이블이 있는것과 없는 것이 혼합된 경우 
* 일반적으로는 일부 데이터에만 레이블이 있음
* 준지도 학습의 알고리즘의 경우는 대부분 지도학습 알고리즘과 비지도학습 알고리즘의 조합으로 구성 

#### 강화학습
* 동적 환경과 함께 상호작용하는 피드백 기반 학습 방법 
* 에이전트가 환경을 관찰하고, 행동을 실행하고 보상 또는 벌점을 받음 
* 에이전트는 이러한 피드백을 통해 자동으로 학습하고 성능을 향상시킴 
* 어떤 지도가 없이 일정한 목표를 수행

#### 온라인 VS 배치 

* 온라인 학습 
    * 적은 데이터를 사용해 미니배치 단위로 점진적으로 학습을 진행 
    * 실시간 시스템이나 메모리 부족의 경우 사용  
* 배치 학습 
    * 전체 데이터를 모두 사용해 오프라인에서 학숩 
    * 컴퓨팅 차원이 풍부한 경우 사용 

#### 사례기반 vs 모델기반 
* 사례기반 학습 
    * 훈련 데이터를 학습을 통해 기억 
    * 예측을 위해 데이터 사이의 유사도 측정 
    * 새로운 데이터와 학습된 데이터를 비교 
* 모델기반 학습 
    * 훈련 데이터를 사용해 모델을 훈련 
    * 훈련된 ㅗ델을 사용해 새로운 데이터를 예측 

### 일반화, 과대 적합, 과소적합 

#### 일반화(Generalization) 
* 훈련된 모델이 처음보는 데이터에 대해 정확히 예측한다면, 이러한 상태를 모델이 **일반화(generalization)** 되었다고 함
* 모델이 항상 일반화 되는것은 아님 

#### 과대적합

#### 과소적합 

#### 모델 복잡도와 데이터셋 크기의 관계 
* 데이터의 다양성이 클수록 더 복잡한 모델을 사용하면 좋은 결과를 얻을수는 있음 
* 일반적으로 더 큰 데이터셋일수록 다양성이 높기 때문에 더 복잡한 모델을 사용할 수 있음 
* 하지만, 같은 데이터를 중복하거나 비슷한 데이터를 모으는 것은 다양성 증가에 도움이 되진 않음 
* 데이터를 더 많이 수집하고 적절한 모델을 만들어서 사용하면 지도 학습을 사용해 놀라운 결과를 얻을 수 있음

### 기계학습이란
* 인공지능의 한 분야 
* 컴퓨터가 학습하 수 있도록 하는 알고리즘과 기술을 개발하는 분야 
* 표현(representation): 데이터의 평가(규칙) 
* 일반화(generalization): 아직 알 수 없는 데이터에 대한 처리(예측) 



##### 기계학습의 차이 
* 통계(통계학) 
    * 표본데이터를 이용하여 모집단에 대해 추론 
* 기계학습(컴퓨터공학) 
    * 전체 데이터를 이용해 개별값 예측 
    * 기존 데이터로 모델을 학습시킨 후 새로운 데이터를 입력했을때 예측값을 알아내기 위한 목적 
* 데이터 마이닝(경영) 
    * 가지고 있는 데이터에서 일정한 패턴이나 의미있는 특성을 발견 
    * 데이터베이스의 지식 발견 부분의 분석절차  
       
       


#### 지도학습(Supervised Learning) 
* 지도학습의 목적 
    * 예측변수를 기반으로 결과를 분류하거나 예측, 즉 결론적으로는 둘 모두 최적의 회귀선을 찾는 과정이다. 
    * 분류: 목표변수가 범주형 변수 
    * 예측: 목표변수가 수치형 변수 

#### 비지도학습(Unsupervised Learning)  
* 데이터가 어떻게 구성되었는지를 알아내는 문제 

#### 강화학습(Reinforcement Learning) 
* 어떤 환경 안에서 정의된 에이전트가 현재의 상태를 인식하여, 선택 가능한 행동들 중 보상을 최대화 하는 행동 혹은 순서를 선택하는 방법
