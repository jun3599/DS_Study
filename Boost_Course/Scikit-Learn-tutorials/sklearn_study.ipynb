# 실습1. classification - Decision Tree   

* 실습에 사용된 소스코드는 사이킷런 공식 홈페이지를 참조했습니다.

## 기본 사용 방식

```python
# 사이킷런에서 tree 모듈을 불러옵니다.
from sklearn import tree
```

```python
# 테스트를 위한 데이터를 준비합니다. 
X = [[0, 0], [1, 1]] # 입력 [0,0]에 대한 y의 값은 0 // 입력 [1,1]에 대한 정답은 1입니다. 
Y = [0, 1]
```

```python
clf = tree.DecisionTreeClassifier() # tree 중 하나인 의사결정나무 모형의 객체를 생성합니다.
clf
```

```python
clf = clf.fit(X, Y) # 생성한 데이터를 모형에 적합합니다. 
```

```python
# 학습된 모형을 통해 정답을 예측합니다. 
clf.predict([[2.,2.]])
```

해당 모형의 경우, 정답 값이 0 또는 1로 결정되어 있기 때문에 예측 값이 1로 나타남을 볼 수 있습니다.   

```python
clf.predict_proba([[2.,2.]])
```

predict_proba의 경우 예측의 확률값을 출력합니다.

## 당뇨병 데이터셋을 통한 심층학습

* pima indian 데이터 셋을 활용하여, 당뇨병 예측을 수행합니다. 
[출처 :: kaggle || pima Indian Diabets Database](https://www.kaggle.com/uciml/pima-indians-diabetes-database) 

* 데이터 구성 
    - pregnancies : 임신횟수  
    - Glucose : 2시간 동안의 경구 포도당 내성 검사에서 혈장 포도당 농도
    - BloodPressure : 이완기 혈압 (mm Hg)
    - SkinThickness : 삼두근 피부 주름 두께 (mm), 체지방을 추정하는데 사용되는 값
    - Insulin : 2시간 혈청 인슐린 (mu U / ml)
    - BMI : 체질량 지수 (체중kg / 키(m)^2)
    - DiabetesPedigreeFunction : 당뇨병 혈통 기능
    - Age : 나이
    - Outcome : 768개 중에 268개의 결과 클래스 변수(0 또는 1)는 1이고 나머지는 0입니다.

### 분석을 위한 사전 준비 

#### 필요 라이브러리 불러오기

```python
# 데이터 분석을 위한 pandas, 수치계산을 위한 numpy
# 시각화를 위한 seaborn, matplotlib.pyplot 을 로드합니다. 

import pandas as pd
import numpy as np # 수치계산을 위한 라이브러리 
import seaborn as sns
import matplotlib.pyplot as plt

# %matplotlib inline
```

#### 데이터셋 로드 

```python
data = pd.read_csv('./datasets/pima_indian_diabetes.csv')
```

```python
print(data.shape)
data.head(10)
```

#### 학습, 예측 데이터셋 나누기 
* 주어진 자료를 순서대로 8:2 비율로 나누어줍니다 

```python
# 8:2 의 비율로 구하기 위해 전체 데이터의 행에서 80% 위치에 해당되는 값을 구해서 split_count 라는 변수에 담습니다.
split_count = int(data.shape[0] * 0.8)
split_count
```

```python
# train, test로 슬라이싱을 통해 데이터를 나눕니다.
train = data[:split_count].copy()
print(train.shape)

test = data[split_count:].copy()
print(test.shape)
```

#### 학습과 예측에 사용될 컬럼명 

```python
# feature_names 라는 변수에 학습과 예측에 사용할 컬럼명을 가져옵니다.
feature_names = train.columns[:-1].tolist()
feature_names
```

#### 예측을 위한 목표변수 컬럼명

```python
# label_name 이라는 변수에 예측할 컬럼의 이름을 담습니다.

label_name = train.columns[-1]
label_name # 이름만 담아둔다. 
```

#### 학습, 예측 데이터 셋 구성하기 

```python
# 학습 세트 만들기 - 사전에 준비한 컬럼명 변수를 이용해 해당 데이터만을 따로 가져와 담습니다.

X_train = train[feature_names]
print(X_train.shape)
X_train.head()
```

```python
# 정답 값을 만들어 줍니다.- 사전에 준비한 컬럼명 변수를 이용해 해당 데이터만을 따로 가져와 담습니다.
y_train = train[label_name]
print(y_train.shape)
y_train.head()
```

```python
# 예측에 사용할 데이터세트를 만듭니다.- 사전에 준비한 컬럼명 변수를 이용해 해당 데이터만을 따로 가져와 담습니다.

X_test = test[feature_names]
print(X_test.shape)
X_test.head()
```

```python
# 예측의 정답값 - 사전에 준비한 컬럼명 변수를 이용해 해당 데이터만을 따로 가져와 담습니다.
y_test = test[label_name]

print(y_test.shape)
y_test.head()
```

### 모형을 통한 예측과 검정

#### 머신러닝 알고리즘 가져오기 

```python
# sklearn 에서 제공하는 의사결정나무 모형을 로드합니다. 
from sklearn.tree import DecisionTreeClassifier # 해당 모형은 분류를 위한 것! 

# 연속형 변수의 예측, 즉 회귀 모형에서는 아래의 모형을 사용합니다. 
# from sklearn.tree import  DecisionTreeRegressor
```

```python
# 모형 생성 및 초기화 
model = DecisionTreeClassifier(random_state=42) # random seed를 임의로 지정하여 결과를 항상 동일하게 고정합니다. 
model
```

#### 학습
* 사전에 준비한 train data를 기반으로 모형을 학습시킵니다. 

```python
model.fit(X_train,y_train)
```

#### 예측 

```python
pred_y = model.predict(X_test) # 모형을 기반으로 추정한 y값을 담아줍니다. 
```

```python
pred_y
```

### 트리알고리즘 분석하기 
* 트리알고리즘의 장점은 시각화의 방법이 간단하고, 시각화를 통해 설명하기 원활함에 있습니다.

트리의 깊이가 깊어질수록 overfitting현상이 발생하는 경향이 있어, 주로 트리의 깊이를 제한하는 방식으로 과적합을 방지합니다. 

#### 피처의 중요도 추출하기 

```python
# 피처 중요도 추출은 feature_importance 메소드를 활용해 간단하게 추출이 가능합니다. 

model.feature_importances_
# 결과값은 전체 데이타(1)을 설명하기 위한 각 변수별 상대적 중요의 수치로 나타나게 됩니다. 
```

### 정확도(Accuracy) 측정하기 

```python
# 우리가 모형을 통해 예측한 값 
pred_y 
```

```python
# 실제 y의 값 
y_test
```

```python
# 기본 컨셉은 다음과 같습니다 
y_test == pred_y
```

* 정확도는 실제값과 예측값의 차이를 의미합니다.
* 따라서 |실제값 - 예측값| 을 기준으로 정확도를 판단합니다. 
    - 해당 값이 0인 경우, 예측의 결과와 실제 값이 동일함을 의미하고, 
    - +1 or -1의 경우에는 절대값이 1 --> 즉 예측과 실제 값이 다름을 의미합니다. 

```python
# 각각의 변수의 정오여부를 알려줍니다. 
abs(y_test - pred_y)
```

```python
# 따라서 전체 값을 더한 후, 총 셈플의 수로 나눈 값은 모형의 정확도를 의미합니다. 
abs(y_test - pred_y).sum() / len(y_test) 
```

즉, 전체중의 28%를 잘못 예측했음을 의미합니다.(정답이면 0을 뱉기 때문)  
따라서, 해당 결정나무의 정확도는 약 71%임을 나타냅니다. 

```python
# 예측의 정확도를 구합니다. 100점 만점 중에 몇 점을 맞았는지 구한다고 보면 됩니다.

diff_count = abs(y_test - pred_y).sum()
print(diff_count)
(len(y_test) - diff_count) / len(y_test) * 100
```

```python
# 위에서 처럼 직접 구할 수도 있지만 미리 구현된 알고리즘을 가져와 사용합니다.

from sklearn.metrics import accuracy_score

accuracy_score(y_test, pred_y) * 100
```

```python
# model 의 score 로 점수를 계산합니다.
model.score(X_test, y_test) * 100
```

값을 요약해 보면 최솟값이 0으로 나오는 값들이 있습니다.

```python
feature_columns
```

```python
# 결측치 여부를 나타내는 데이터프레임을 만듭니다.
# 0값을 결측치라 가정하고 정답(label, target)값을 제외한 컬럼에 대해 
# 결측치 여부를 구해서 df_null 이라는 데이터프레임에 담습니다.

cols =feature_columns[1:] # 출산 횟수의 경우, 0이 결측임을 나타내지는 않기 때문에 제외합니다. 

df_null = data[cols].replace(0, np.nan) # 직접적으로 데이터에 변화를 주는것이 아닌, 따로 처리합니다. 
df_null = df_null.isnull()
df_null.sum() # 0 을 결측치로 보았을때, 아래와 같은 결측 수를 확인할 수 있습니다. 
```

```python
# 결측치의 갯수를 구해 막대 그래프로 시각화 합니다.
df_null.sum().plot.barh()
```

```python
# df_null 데이터는 True False 값으로 구성되어 있습니다. 
# 따라서 아래의 방식으로 각 데이터에서 결측의 비율을 확인 할 수 있습니다. 
df_null.mean() * 100
```

 ##### 반복문을 통해 subplot 그리기 

```python
# 반복문의 사용을 위한 사전 준비 
# 컬럼의 수 만큼 for 문을 만들어서 서브플롯으로 시각화를 합니다.
cols = data.columns[:-1].tolist()
cols
```

```python
# distplot 으로 서브플롯을 그립니다.

# 여러개의 subplot 객체를 포함하는 fig 객체 생성 
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 15)) # 총 크기는 15*15 이며, 3개의 열과 3개의 행 --> 9개의 총 공간을 할당합니다. 

# enumerate의 경우, 리스트를 가지고 반복문을 순회할때, 인덱스와 값을 같이 가져옵니다. 
for i, col_name in enumerate(cols):
    row = i // 3 # 3개가 채워지기 전까지 같은 행을 유지 
    col = i % 3 # 3개 안에서 1칸씩 이동 
    sns.distplot(data[col_name], ax=axes[row][col]) # ax 옵션으로 위치할 행열 정보를 전달합니다. 
```

위의 그림 또한, 레이블을 고려하지 않기 때문에, 발병 여부에 따른 데이터는 보기 어렵습니다.  


```python
# 따라서, 발병 여부로 나누어 놓은 데이터 두개를 각각 동시에 그려 분포의 차이를 살펴봅니다. 
# 모든 변수에 대한 distplot을 그려 봅니다.

fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(15, 15))

for i, col_name in enumerate(cols[:-1]):
    row = i // 2
    col = i % 2
    # 같은 영역에 각각 2개(0,1)의 데이터를 같이 그려줍니다. 
    sns.distplot(df_0[col_name], ax=axes[row][col]) 
    sns.distplot(df_1[col_name], ax=axes[row][col])
    
```
