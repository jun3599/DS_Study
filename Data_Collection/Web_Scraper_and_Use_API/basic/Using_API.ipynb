* 기본적인 접근순서는 위와 같습니다. 
* 다만, 여기에서는 xml전용 parser인 ElenentTree를 활용하여 스크래핑을 진행합니다. 

```python
from urllib.request import Request, urlopen
from urllib.parse import urlencode, quote_plus
from xml.etree import ElementTree
import pandas as pd 

skey = 'I07IIoZaK0NQU%2B9%2BNHLYQwVIDKnPICBgIW8bjz3jYqUebxCyvGKsPdn5wueXFXwbOKQ8wpCrfTSfri48jO6H3A%3D%3D'

url = 'http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19SidoInfStateJson'
queryParams = '?' + 'ServiceKey=' + skey + '&'+urlencode({quote_plus('pageNo') : '1', quote_plus('numOfRows') : '10', quote_plus('startCreateDt') : '20200120', quote_plus('endCreateDt') : '20201231' })

# 해당 페이지에 접근하는 부분 
request = Request(url + queryParams)
request.get_method = lambda: 'GET'
# 해당 페이지를 읽어들이는 부분 
response_body = urlopen(request).read()
print(response_body)

root = ElementTree.fromstring(response_body)

covid_19 = pd.DataFrame()
for item in root.iter('item'):
    try:
        item_dict = {}
        item_dict['createDt'] = item.find('createDt').text
        item_dict['deathCnt'] = item.find('deathCnt').text
        item_dict['defCnt'] = item.find('defCnt').text
        item_dict['gubun'] = item.find('gubun').text
        item_dict['gubunCn'] = item.find('gubunCn').text
        item_dict['gubunEn'] = item.find('gubunEn').text
        item_dict['incDec'] = item.find('incDec').text
        item_dict['isolClearCnt'] = item.find('isolClearCnt').text
        item_dict['isolIngCnt'] = item.find('isolIngCnt').text
        item_dict['localOccCnt'] = item.find('localOccCnt').text
        item_dict['overFlowCnt'] = item.find('overFlowCnt').text
        item_dict['qurRate'] = item.find('qurRate').text
        item_dict['seq'] = item.find('seq').text
        item_dict['stdDay'] = item.find('stdDay').text
        #item_dict['updateDt'] = item.find('updateDt').text

        covid_19 = covid_19.append(item_dict,ignore_index=True)
    except:
        pass
```
