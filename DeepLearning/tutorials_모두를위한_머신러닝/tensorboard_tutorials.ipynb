# Tensorboard Tutorials 
* Tensorboard는 tensorflow에서 발생한 학습 로그를 표시하거나 디버깅을 하기 위한 도구로, 주로 시각화를 진행할때 사용한다. 

![tensorboard.png](https://tutorials.pytorch.kr/_static/img/tensorboard_scalar_runs.png)
* tensorboard는 5단계를 거치면 손쉽게 사용이 가능합니다. 
![tuto.png](https://img1.daumcdn.net/thumb/R800x0/?scode=mtistory2&fname=https%3A%2F%2Ft1.daumcdn.net%2Fcfile%2Ftistory%2F9964214B5B303C9120)  
1. tf그래프에서, 우리가 어떤 값(텐서)를 추적하고 싶은지 정합니다. 
   이는, 제공하는 함수를 사용하여 구현이 가능합니다.  
   - tf.summary.histogram("이름", 값(텐서))
   - tf.summary.scalar("이름", 값(텐서)) 
2. 추적을 심어둔 summary들을 개별 값이 아닌 한번에 다루기 위해, merge all 시켜줍니다. 
    - summary = tf.summary.merge_all() 

3. 일종의 file writer를 생성합니다. 
    - 이는 해당 정보를 임의의 세션에 올려두는 작업을 의미하며, 이와 동시에 저장을 할 위치의 경로를 입력해줍니다. 
    - writer = tf.summary.FileWriter('./logs')
    - writer.add_graph(sess.graph) # 세션에 그래프를 추가합니다. 
4. summary를 실행해줍니다. 
    - summary 또한 일종의 텐서이기 때문에 실행해줍니다. 
    - s, _ = sess.run([summary, optimizer], feed dict = feed_dict)
    - writer.add_summary(s, global_step = global_step) # 위의 코드의 실행으로 발생한 실제 결과를 파일에 추가하는 부분 

5. 저장결과 실행 
$ tensorboard -logdir= ./logs
